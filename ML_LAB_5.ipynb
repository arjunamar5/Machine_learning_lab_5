```python
#A2
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score
from math import sqrt

# Load the dataset
file_path = r'C:\Users\year3\Downloads\DCT_withoutduplicate 5.csv'  # Update this path if needed
df = pd.read_csv(file_path)

# Split into training and test sets (80-20 split as an example)
train_size = int(0.8 * len(df))
df_train = df[:train_size]
df_test = df[train_size:]

# Select the first attribute (column 0) as the feature and LABEL as the target
X_train = df_train[[df.columns[0]]]  # Using column 0 as the feature
y_train = df_train['LABEL']  # Using LABEL as the target variable

X_test = df_test[[df.columns[0]]]  # Same feature for the test set
y_test = df_test['LABEL']  # Same target for the test set

# Train a linear regression model
reg = LinearRegression().fit(X_train, y_train)

# Predictions on training data
y_train_pred = reg.predict(X_train)

# Predictions on test data
y_test_pred = reg.predict(X_test)

# Calculate metrics for the training set
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = sqrt(mse_train)
mape_train = mean_absolute_percentage_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

# Calculate metrics for the test set
mse_test = mean_squared_error(y_test, y_test_pred)
rmse_test = sqrt(mse_test)
mape_test = mean_absolute_percentage_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

# Print the results
print(f"Training Set: MSE = {mse_train}, RMSE = {rmse_train}, MAPE = {mape_train}, R2 = {r2_train}")
print(f"Test Set: MSE = {mse_test}, RMSE = {rmse_test}, MAPE = {mape_test}, R2 = {r2_test}")
```

```python
#A3
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score
from math import sqrt
from sklearn.cluster import KMeans

# Load the dataset
file_path = r'C:\Users\year3\Downloads\DCT_withoutduplicate 5.csv'  # Ensure the path is correct
df = pd.read_csv(file_path)

# Split into training and test sets (80-20 split as an example)
train_size = int(0.8 * len(df))
df_train = df[:train_size]
df_test = df[train_size:]

# A1: Linear Regression with Multiple Attributes
# Use all attributes except 'LABEL' for the features
X_train = df_train.drop(columns=['LABEL'])
y_train = df_train['LABEL']

X_test = df_test.drop(columns=['LABEL'])
y_test = df_test['LABEL']

# Train the linear regression model
reg = LinearRegression().fit(X_train, y_train)

# Predictions on training data
y_train_pred = reg.predict(X_train)

# Predictions on test data
y_test_pred = reg.predict(X_test)

# Calculate metrics for the training set
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = sqrt(mse_train)
mape_train = mean_absolute_percentage_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

# Calculate metrics for the test set
mse_test = mean_squared_error(y_test, y_test_pred)
rmse_test = sqrt(mse_test)
mape_test = mean_absolute_percentage_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

# Output the results for Linear Regression
print("Linear Regression - Training Set: MSE =", mse_train, ", RMSE =", rmse_train, 
      ", MAPE =", mape_train, ", R2 =", r2_train)
print("Linear Regression - Test Set: MSE =", mse_test, ", RMSE =", rmse_test, 
      ", MAPE =", mape_test, ", R2 =", r2_test)

# A2: K-Means Clustering with Multiple Attributes
# Use all attributes except 'LABEL' for clustering
X = df.drop(columns=['LABEL'])

# Apply k-means clustering with 2 clusters and specify n_init as an integer
kmeans = KMeans(n_clusters=2, random_state=0, n_init=10).fit(X)

# Retrieve cluster labels and cluster centers
labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

# Output the results for K-Means Clustering
print("K-Means Clustering - Cluster Labels:", labels)
print("K-Means Clustering - Cluster Centers:\n", cluster_centers)
```

```python
#A5
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Load the dataset
file_path = r'C:\Users\year3\Downloads\DCT_withoutduplicate 5.csv'  # Ensure the path is correct
df = pd.read_csv(file_path)

# Use all attributes except 'LABEL' for clustering
X = df.drop(columns=['LABEL'])

# Apply k-means clustering with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10).fit(X)

# Calculate the Silhouette Score
sil_score = silhouette_score(X, kmeans.labels_)

# Calculate the Calinski-Harabasz Score
ch_score = calinski_harabasz_score(X, kmeans.labels_)

# Calculate the Davies-Bouldin Index
db_index = davies_bouldin_score(X, kmeans.labels_)

# Output the clustering evaluation metrics
print(f"Silhouette Score: {sil_score}")
print(f"Calinski-Harabasz Score: {ch_score}")
print(f"Davies-Bouldin Index: {db_index}")
```

```python
#A6
import numpy as np
import pandas as pd  # Import pandas for reading the CSV file
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load dataset from the CSV file
file_path = r'C:\Users\year3\Downloads\DCT_withoutduplicate 5.csv'
df = pd.read_csv(file_path)

# Assuming the data has numerical features, let's use them for clustering
# If your dataset has specific columns, replace df.values with df[['column1', 'column2']].values
X = df[['LABEL']]

# Range of k values to test
k_values = range(1, 11)

# Lists to store evaluation metrics
inertia_values = []
silhouette_scores = []

# Perform k-means clustering for each k value
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)
    
    if k > 1:  # Silhouette score is not defined for k=1
        silhouette_avg = silhouette_score(X, kmeans.labels_)
        silhouette_scores.append(silhouette_avg)
    else:
        silhouette_scores.append(float('nan'))  # Placeholder for k=1

# Plot inertia vs k
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(k_values, inertia_values, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Inertia vs. Number of Clusters')
plt.grid(True)

# Plot silhouette score vs k
plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load dataset from the CSV file
file_path = r'C:\Users\year3\Downloads\DCT_withoutduplicate 5.csv'
df = pd.read_csv(file_path)

# Assuming the data has a numerical column named 'LABEL', use it for clustering
# Replace 'LABEL' with the actual column name(s) you want to use
X = df[['LABEL']].values  # Use .values to convert to numpy array

# Initialize list to store distortion values
distortions = []

# Perform k-means clustering for each k value from 2 to 19
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(X)
    distortions.append(kmeans.inertia_)

# Plot distortion values to use the elbow method
plt.figure(figsize=(8, 5))
plt.plot(range(2, 20), distortions, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion (Inertia)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```
